```python
from glm.spk_train import Spike_train as SPK
from mle.inference import Maximum_likelihood_estimator as MLE
import numpy as np
import matplotlib.pyplot as plt
%autosave 1
```

```python
# simulate spike train
N, Nt = 2, 2000000
spk_train = SPK(N=N, Nt=Nt, dt=0.05, p=0.3, weight_factor=0.9)
```

```python
spk_train.weight_matrix = np.array([[-0.        ,  0.29058073],
                                    [-0.2        , -0.        ]])
```

```python
spk_train.simulate_poisson()
```

```python
# mle
observed_neurons = [0, 1]
mle = MLE(filter_length=100, dt=0.1, basis_order=[0, 1, 2], observed=observed_neurons, tau=1)
mle.plot_inferred(spk_train.spike_train, spk_train.weight_matrix)
```
